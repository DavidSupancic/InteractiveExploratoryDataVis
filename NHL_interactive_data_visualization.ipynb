# Interactive Exploratory Data Visualization of NHL Players

## Data Loading

```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('NHL_data/NHL_vis.csv', index_col=0)
df.head()
```

```python
df.shape
```

## Interactive Exploratory Data Visualization

```python
fig = px.scatter(df, x="Age", y="G", 
                 color="Pos",
                 hover_data=["Player", "Tm"],
                 title="Goals vs Age by Position")
fig.show()
```
