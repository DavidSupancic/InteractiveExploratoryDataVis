# Interactive Exploratory Data Visualization of NHL Players

## Data Loading

```python
import pandas as pd
import plotly.express as px

import warnings
warnings.filterwarnings('ignore', category=FutureWarning)
```

```python
df = pd.read_csv('NHL_data/NHL_vis.csv', index_col=0)
df.head()
```

```python
df.shape
```

Let's create Goals/Game column.

```python
df['Goals/Game'] = df['G'] / df['GP']
```

### Scatter Matrix

```python
# Interactive Scatter Matrix

fig = px.scatter_matrix(df, 
                       dimensions=['Age', 'GP', 'G', 'plusminus'],
                       color='Pos',
                       title='NHL Players Scatter Matrix',
                       hover_data=['Player', 'Tm', 'Season'],
                       color_discrete_sequence=px.colors.qualitative.Set3)

fig.update_layout(
    width=1400,
    height=1600
)

fig.show()
```
